```python
from openai import OpenAI
```

```python
client = OpenAI(api_key=openai_key)
```

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role":"system", "content":"You are an AI model that takes instructions and generates output."},
        {"role":"user", "content":"Who won T20 world cup?"}
        ]
)
print(response.choices[0].message.content)
```

```python
def chat(message):
  response = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role":"system", "content":"You are an AI model that takes instructions and generates output."},
        {"role":"user", "content": f"{message}"}
        ]
  )
  temp = response.choices[0].message.content
  return temp
```

```python
chat("what is the capital of Gujarat and Ahmedabad?")
```

```python
chat("Where should we go this December?")
```
